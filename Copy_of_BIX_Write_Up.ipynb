Looking at the above decay of firing rates, the 50% firing rate shows the most striking model difference. The integration circuit becomes almost completely stagnant, which can be explained by the neuron's inability to categorize the stimuli. They do not have a fast enough response time to compete. The sensory model at 50% firing capacity also sees a severe drop-off in activity, with only more central neurons responding to the stimuli. This is likely because, with such a low rate of firing, the signal does not have time to make it to every neuron in the system. This pattern is consistent with the rastergrams seen for the original, 90% firing rate, and 75% firing rate sensory circuits as well: the middle of the rastergram is quite active, but activity falls off at the edges. 
We don't see much of a difference between the control, 90% firing rates, and 75% firing rates.

## Section 2 - EXTERNAL CONNECTIVITY
In this section we alter the external connectivity of neurons in both the sensory and integration circuits by reducing them to 90% connectivity and 75% connectivity in order to show that lower connectivity will lead to less spiking.

In both the sensory and integration circuits, we manipulated external connectivity by changing the connection probability and the weight of external synapses to both inhibitory and excitatory synapses.

Section 3 - INTERNAL CONNECTIVITY

In this section, we reduce internal connectivity in both the sensory and integration circuits to 90% and 75%, expecting to see similar results to our reduction of external connectivity.

In the integration circuit, we manipulate internal connectivty by adjusting the relative recurrent synaptic strength across both populations, the weight of AMPA and NDMA synapses between excitatory neurons, the weight of excitatory to inhibitory synapses, the weight of excitatory synapses, the weight of inhibitory to inhibitory synapses, and the transmission delay of recurrent excitatory to inhibitory connections. 

In the sensory circuit, we manipulate internal connectivity by reducing the connection probability, the relative synaptic strenght of connections between the neuron populations, the weight of excitatory to inhibitory synapses, the weight of excitatory synapses, the weight of inhibitory to inhibitory synapses, and the weight of excitatory to excitatory synapses.

Decreasing internal connectivity did not perform as expected at all — while more neurons spiked in both the integration circuit and sensory circuit rastergrams, the integration circuit saw significantly a significantly lower but more uniform response rates, while the sensory circuit saw signficantly higher and more uniform response rates. In saying 'more uniform,' we mean that both neuron populations (red and blue) follow similar patterns; in the control, and under conditions where we changed both the firing rates and external connectivity, both neuron populations seemed to remain, generally, in competition with each other. ## Conclusions
Going into this, I was hihgly suspicious of the model's capabilities, as Wimmer's paper scrapped the data from one of its subjects in what I was assumed was an effort to avoid data that didn't align with it's model. I found, however, the model responded very well to lowered neuronal thresholds; while it struggled with some of the extremely lowered bounds (such as 50% firing rate), it accurately portrayed all of neural activity according to the thresholds I set. 

It was also interesting to see how similarly the integration circuit and sensory circuit responded to the changed conditions — they followed similar trends under both adjusted firing rate and adjusteed external connection, although they differed greatly under different internal connections. The sensory circuit's increased response to decreased internal connections is likely more reactive because it is more sensitive; neurons exhibit a large response variability, so the decreased internal connections mattered less, as seen when comparing the rate of response in the 75% internal connection to the control. 

As expected, changing external connectivity yielded the most dramatic response in the model, likely because it actively inhibited the neuron's ability to communicate with each other.
